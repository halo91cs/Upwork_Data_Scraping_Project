```python
### We got all doctors' data with requested specialties from bcbs.com ###
### Extract lastName - firstName - middleInitial - address - suite - 
        ##  city - county - state - zipcode - website - phone - fax - MD/DO
import pandas as pd
import json
import os
from csv import writer
import numpy as np
```

```python
path_to_json = '/path/to/json/files/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
```

```python
## We create a csv for each json file to store desired column names in that csv file ##
def create_csv(json_file):
    file_name = json_file + ".csv"
    col_names = ["Last Name", "First Name", "Middle Name", "Degree Label", "Specialty", "Street Address", "Suite", "City", "State", "County", "Zipcode", "Phone Number", "Fax Number", "Website"]
    df_frame =  pd.DataFrame(columns=col_names)
    df_frame.to_csv(file_name)
    
for i in json_files:
    create_csv(i)
```

```python
def append_list_as_row(file_name, list_of_elem):
    # Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)
```

```python
def processJsonFiles(json_file):    
    # Reading the json as a dict
    with open(json_file) as json_data:
        data = json.load(json_data)
    
    file_name = json_file + ".csv"
    for i in (data['results'][0]['providers']):
        all_data = []
        try:
            last_name = i['lastName']
        except:
            last_name = ""
            
        try:
            first_name = i['firstName']
        except:
            first_name = ""
        
        try:
            middle_name = i['middleName']
        except:
            middle_name = ""
            
        try:
            degree = i['credentialDegreeLabel']
        except:
            degree = ""
            
        try:
            specialty = i['bestLocation']['specialties'][0]['specialtyName']
        except:
            specialty = ""
            
        try:
            addressLine1 = i['bestLocation']['addressLine1']
        except:
            addressLine1 = ""
            
        try:
            suite = i['bestLocation']['addressLine2']
        except:
            suite = ""
            
        try:
            city = i['bestLocation']['city']
        except:
            city = ""
            
        try: 
            state = i['bestLocation']['state']
        except:
            state = ""
        
        try: 
            county = i['bestLocation']['county']
        except:
            county = ""
        
        try:
            zip_code = i['bestLocation']['zip']     
        except:
            zip_code = ""
        
        try:
            phone_no = i['bestLocation']['officePhoneNumber']
        except:
            phone_no = ""
        
        try:
            fax_no = i['bestLocation']['faxNumber']
        except:
            fax_no = ""
            
        try:
            website = i['bestLocation']['locationWebsite'][0]['url']
        except:
            website = ""
        
        all_data.append("")
        all_data.append(last_name)
        all_data.append(first_name)
        all_data.append(middle_name)
        all_data.append(degree)
        all_data.append(specialty)
        all_data.append(addressLine1)
        all_data.append(suite)
        all_data.append(city)
        all_data.append(state)
        all_data.append(county)
        all_data.append(zip_code)       
        all_data.append(phone_no)
        all_data.append(fax_no)
        all_data.append(website) 
        
        append_list_as_row(file_name, all_data)     
```

```python
for i in json_files:
    processJsonFiles(i)
```

```python
bom = pd.read_csv('BoardOfMedicineFinal.csv')
boom = pd.read_csv('finalBoard_of_Ost_Phy.csv')
bom.drop(['Indexes', 'LicenseNumber','Degree', 'City', 'Status', 'Link', 'Unnamed: 8'], axis=1, inplace=True)
boom.drop(['Unnamed: 0', 'License', 'Profession', 'City', 'License Status', 'Urls'], axis=1, inplace=True)
frames = [bom, boom]
allBoms = pd.concat(frames)
allBoms.reset_index(drop=True, inplace=True)
```

```python
allBoms['Name'] = allBoms['Name'].apply(lambda row: row.lower())
allBoms['Name'] = allBoms['Name'].apply(lambda row: row.replace(',',''))
allBoms['Name'] = allBoms['Name'].apply(lambda row: row.strip())
allBoms['Middle'] = allBoms['Name'].apply(lambda row: row.split(' ')[2] if len(row.split(' ')) == 3 else "")
allBoms['Name'] = allBoms['Name'].apply(lambda row: row.replace(row.split(' ')[2], row.split(' ')[2][0]) if len(row.split(' ')) == 3 else row)
allBoms['Email'] = allBoms['Email'].apply(lambda row: row.replace(row, 'NA') if 'Section' in str(row) else row)
allBoms['Name'] = allBoms['Name'].apply(lambda row: " ".join(row.split()))

allBomsDict = dict(zip(allBoms['Name'], allBoms['Email']))
```

```python
path_to_csv = '/path/to/csv/files/'
csv_files = [pos_csv for pos_csv in os.listdir(path_to_csv) if 'json.csv' in pos_csv]
```

```python
## Get doctors for each specialty ## 
dfs=[]
for i in csv_files:
    df = pd.read_csv(i)
    df['Last Name'] = df['Last Name'].apply(lambda row: " ".join(row.split()))
    df['First Name'] = df['First Name'].apply(lambda row: " ".join(row.split()))
    df['Full Name'] = df.apply(lambda row: row['Last Name'] + " " + row['First Name'] + " " + str(row['Middle Name']) if str(row['Middle Name']) != "nan" else row['Last Name'] + " " + row['First Name'], axis=1)
    df['Full Name'] = df['Full Name'].apply(lambda row: row.lower())
    df['Full Name'] = df['Full Name'].apply(lambda row: row.strip())
    dfs.append(df)
```

```python
for j in dfs:
    j['Email'] = ""
    k=0
    for i in j['Full Name']:
        try:
            j['Email'].iloc[k] = allBomsDict[i]                
        except:
            keys = [] ## it will hold all keys that satisfy following conditions
            for key in allBomsDict:                
                all_names = i.split(' ') ## austin mitchell b will be [austin, mitchell, b]
                all_names_len = len(all_names)
                if all_names_len == 3: ## if length is 3 it means there is a middle name
                    if (all_names[0] in key) and (all_names[1] in key) and (all_names[2] in key):
                        keys.append(key)
                else: ## if length is not 3. if it is not 3 it must be 2 meaning there is no middlename
                    if (all_names[0] in key) and (all_names[1] in key):
                        keys.append(key)
                
            ## if there is no key that meets the above conditions then there is no email record for that person 
            if len(keys) == 0:
                j['Email'].iloc[k] = "NA"
                
            else:
                ## find similarity between the original name and all keys ##
                ratios = []
                for p in keys:
                    ratios.append(similar(p, i))

                maxInd = np.argmax(ratios)
                j['Email'].iloc[k] = allBomsDict[keys[maxInd]]           
                
        k+=1
```
